**Task-3 Outliner Detection**

```python
#question1 - 2
def outliers(data,column):
  q1=data[column].quantile(0.25)
  q3=data[column].quantile(0.75)
  Iqr=q3-q1
  lower_bound=q1-1.5*Iqr
  upper_bound=q3+1.5*Iqr
  return data[(data[column]<lower_bound) | (data[column]>upper_bound)]
outliers_gdp=outliers(south_asian_df,'Log GDP per capita')
outliers_score=outliers(south_asian_df,'score')
print("\nOutliers based on GDP per capita:")
print(outliers_gdp)
print("\nOutliers based on score:")
print(outliers_score)
```

```python
#question-3
plt.figure(figsize=(12,6))
sns.scatterplot(
    x='Log GDP per capita',
    y='score',
    data=south_asian_df,
    color='Green',
    label='Non Outliers'
)
sns.scatterplot(
    x=outliers_gdp['Log GDP per capita'],
    y=outliers_gdp['score'],
    data=outliers_gdp,
    color='orange',
    label='Outliers'
)
plt.title('Outliers detection')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

**Outliner Detection**

```python
#Outlier Detection
def detect_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

sa_outliers = detect_outliers(south_asian_df, 'score')
me_outliers = detect_outliers(middle_east_df, 'score')

# Scatter plot for outliers in South Asia
sns.scatterplot(x="Log GDP per capita", y="score", data=south_asian_df, color="blue", label="Non-Outliers")
sns.scatterplot(x=sa_outliers["Log GDP per capita"], y=sa_outliers["score"], color="red", label="Outliers")
plt.title("Outliers Detection in South Asia")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()
plt.show()
print("\n")

# Scatter plot for outliers in Middle East
sns.scatterplot(x="Log GDP per capita", y="score", data=middle_east_df, color="blue", label="Non-Outliers")
sns.scatterplot(x=me_outliers["Log GDP per capita"], y=me_outliers["score"], color="red", label="Outliers")
plt.title("Outliers Detection in Middle East")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()
plt.show()
print("\n")
```
